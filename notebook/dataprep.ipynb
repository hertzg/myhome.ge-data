```python
products = pd.read_json('../data/products.json')
maklers = pd.read_json('../data/maklers.json')
floors = pd.read_json('../data/floors.json')
users = pd.read_json('../data/users.json')
points = pd.read_json('../data/mapData_points.json')
```

```python
#products.head()
columns = filter(lambda col: not col.endswith('_json'), products.columns) 
clean_products = products[columns].set_index('product_id')
clean_products.head()
```

```python
maklers.head()
```

```python
#floors.head()
clean_floors = floors[['developer_flat_id', 'product_id','floor','price','price_value','currency_id']].set_index('product_id')
clean_floors.head()
```

```python
#users.head()
clean_users = users[['user_id', 'username', 'gender_id']].set_index('user_id')
clean_users.head()
```

```python
#points.head()
clean_points = points[['product_id', 'title', 'owner_type', 'href']]
clean_points = clean_points \
    .join(points['price'].apply(pd.Series)) \
    .join(points['center'].apply(pd.Series)) \
    .set_index('product_id')
clean_points.head()
```

```python
full = clean_products \
    .join(clean_floors, rsuffix='_floor') \
    .join(clean_users, on='user_id', rsuffix='_user') \
    .join(clean_points, rsuffix='_point') \
    .drop_duplicates()
full.head()
```
